```python
from keras.models import load_model
from keras.preprocessing.image import img_to_array
import numpy as np
import os
import cv2
import numpy as np
from os import listdir
from os.path import isfile, join
import re
import matplotlib.pyplot as plt
```

```python
model = load_model('test1.h5')
```

```python
image = cv2.imread('./images/happy.jpg')
face_cascade = cv2.CascadeClassifier('./models/haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(image, 1.1, 4)
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h),(0, 0, 255), 2)      
    faces = image[y:y + h, x:x + w]
plt.imshow(faces)
plt.show()
```

```python
test_image = faces.copy()
test_image = cv2.resize(test_image, (48, 48), interpolation = cv2.INTER_LINEAR)
test_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
test_image = img_to_array(test_image)
test_image = np.reshape(test_image, (1,48, 48, 1))
test_image = test_image * 1./255

predictions = model.predict(test_image).flatten()
```

```python
predictions = predictions
result = np.argmax(predictions)
```
